## Traits from existing sources

Read trait info into a data frame.

```python
qry = """
SELECT code, name, description, value_type, life_stage,
life_history_process, category_vocabulary, method_vocabulary, priority
FROM litrev.trait_info ;
"""
res = dbquery(qry,db_v1_1)
data = pd.DataFrame(res)
trait_info=data.rename(columns={0:"Trait code", 1:"Trait name", 2:"Description", 4:"Life stage", 5:"Life history process",  3:"Value type",6:"category_vocabulary",7:"method_vocabulary", 8:"priority"})
```
