## Set up
Load libraries

```python
from pathlib import Path
import os
from configparser import ConfigParser
import psycopg2
from psycopg2.extras import DictCursor
import pandas as pd
import pyprojroot
from datetime import date
```

Load functions from `lib` folder, we will use a function to read db credentials and one for database queries:

```python
from lib.parseparams import read_dbparams
from lib.firevegdb import dbquery
```

Find project root directory

```python
repodir = pyprojroot.find_root(pyprojroot.has_dir(".git"))
```

Read database credentials from `database.ini` file

```python
dbparams = read_dbparams(repodir / 'secrets' / 'database.ini', section='aws-lght-sl')
```

### Fire history

```python
qry="""
SELECT site_label,count(*),MIN(earliest_date),MAX(latest_date)
FROM form.field_site
LEFT JOIN form.fire_history
USING (site_label)
GROUP BY site_label
"""
res = dbquery(qry,dbparams)
colnames=['site_label','fire_records','min_date','max_date']
fire_history = pd.DataFrame(res,columns=colnames)
```

```python
fire_history.fire_records.sum()
fire_history
```
